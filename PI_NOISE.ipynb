```python
class RAF:
  def __init__(self, path):
    self.raf_file = self.open(path)

  def open(self, raf_path):
    try:
      return open(raf_path, 'r+b')

    except:
      RAF_FILE = open(raf_path, 'wb')
      RAF_FILE.close()
      return open(raf_path, 'r+b')

  def close(self):
    self.raf_file.close()

  def get_data(self, index, buffer):
    self.raf_file.seek(index)
    data = self.raf_file.read(buffer)
    return data.decode('utf-8').replace('\x00', '')

  def set_data(self, index, buffer, data):
    buffer = buffer if buffer else len(data)
    self.raf_file.seek(index)
    self.raf_file.write(data.encode('utf-8')[:buffer])

```

```python
import numpy as np

class PiNoise:
  def __init__(self, noise_size=64, m_dim=1, data_path=''):
    self.PI_NUMBERS = self.get_pi_data_file(data_path)
    self.noise_size = noise_size
    self.m_dim = m_dim

    self.noise_area = noise_size ** 2

  def get_pi_data_file(self, data_path):
    data_path = data_path if data_path else '/content/sample_data/pi_hex_1b.raf'
    return RAF(data_path)

  def get_noise_from(self, X, Y):
    M = self.m_dim
    N_AREA = self.noise_area
    buffer = self.noise_area

    if X > M:
      raise ValueError("O valor de X ({}) nao pode ser superior a:{}".format(X, M))

    index = ( ( ( Y - 1 ) * M + X ) * N_AREA ) - N_AREA

    noise_raw = self.PI_NUMBERS.get_data(index, buffer)

    noise_norm = np.array([int(number, 16) / 15 for number in noise_raw])

    return noise_norm.reshape(self.noise_size, self.noise_size)


```

```python
PI_NOISE = PiNoise(10, 10)

for index, number in enumerate(PI_NOISE.get_noise_from(2,1), 1):
  print(index, len(number))
```
